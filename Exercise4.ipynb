```python
def is_even(x):
    return x % 2 == 0
assert is_even(0) == True
assert is_even(101) == False
assert is_even(982) == True
assert is_even(-5) == False
assert is_even(-8) == True
print(is_even(0))
print(is_even(101))
print(is_even(982))
print(is_even(-5))
print(is_even(-8))
```

```python
import sympy as sp

def solve_second_order_ode():
    x = sp.symbols('x')
    y = sp.Function('y')

    print("Solve a second-order ODE of the form:")
    print("a*y'' + b*y' + c*y = f(x)\n")

    a = float(input("Enter coefficient a: "))
    b = float(input("Enter coefficient b: "))
    c = float(input("Enter coefficient c: "))

    fx_input = input("Enter f(x) (e.g., 0, sin(x), x**2, exp(x)): ")
    f = sp.sympify(fx_input)

    ode = sp.Eq(a*y(x).diff(x, 2) + b*y(x).diff(x) + c*y(x), f)

    general_solution = sp.dsolve(ode, y(x))
    sol = general_solution.rhs

    print("\nGeneral Solution:")
    print(sol)

    print("\nEnter auxiliary (initial) conditions:")
    x0 = float(input("Enter x0: "))
    y0 = float(input("Enter y(x0): "))
    dy0 = float(input("Enter y'(x0): "))

    dy = sp.diff(sol, x)

    constants = sp.solve(
        [
            sol.subs(x, x0) - y0,
            dy.subs(x, x0) - dy0
        ]
    )

    particular_solution = sol.subs(constants)

    print("\nFinal Solution after applying conditions:")
    print(particular_solution)


solve_second_order_ode()
7
```
